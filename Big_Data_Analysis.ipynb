#======================== DARI ASSESING DATA =======================#

Setelah dilakukaknya assesing data atau menilai kualitas dari seluruh data yang akan digunakan terdapat informasi yang dapat diambil seperti berikut:

    TIPE DATA        --->  Terdapat kesalahan tipe data pada kolom reservation_status_date

    MISSING VALUES   ---> 4 variable yang terdapat nilai null, yaitu kolom [children, country, agent, company]

    DUPLICATE DATA   ---> Tidak terdapat duplikasi data  


> Solusi untuk missing value

Dari informasi diatas terdapat masalah pada kolom **children**, **country**, **agent**, dan **company**. Dimana keempat kolom terdapat missing value atau Nan value. Terdapat tiga metode untuk mengatasi missing value yaitu *dropping*, *imputation*, dan *interpolation*. Teknik ***Dropout*** akan dilakukan dikarenakan sebagai berikut:
- **Jumlah Data yang Hilang Sedikit**: Karena hanya sedikit yang memiliki nilai NaN, menghapusnya tidak akan mengurangi banyak data dan kemungkinan besar tidak akan mempengaruhi analisis.
- **Kehilangan Data Tidak Signifikan**: Data pada baris-baris yang hilang tidak mengandung informasi yang krusial, penghapusan baris akan lebih mudah dan lebih cepat daripada melakukan imputation yang berpotensi memperkenalkan bias.
- **Variabel tidak relevan**:
Karena variabel agent dan company tidak relevan pada variabel pembatalan pemesanan maka variabel tersebut akan di drop

> Solusi untuk tipe data

- Tipe data pada kolom **reservation_status_date** perlu diubah ke *datetime*.

> Solusi untuk duplikasi data

- akan dilakukan drop data, karena agar tidak ada redudansi data

## Pertanyaan: Berapa banyak pesanan yang diterima dan dibatalkan dari masing-masing tipe hotel?

```python
# Set the background color of the figure
plt.figure(figsize=(8, 4), facecolor='white')

# Create the countplot
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette="Set2")

# Customize legend location
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))

# Set plot title and axis labels
plt.title('Reservation Status in Different Hotels', size=20, color='Black')
plt.xlabel('Hotel', color='Black')
plt.ylabel('Number of Reservations', color='Black')

# Customize legend labels
plt.legend(['Not Cancelled', 'Cancelled'])

# Annotate bars with data labels for exact counts
for p in ax1.patches:
    ax1.annotate(f'{int(p.get_height())}',
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', xytext=(0, 10),
                 textcoords='offset points', fontsize=10, color='black')

# Calculate and print the counts per hotel for Cancelled/Not Cancelled
counts = df.groupby(['hotel', 'is_canceled']).size().unstack(fill_value=0)
print("\nJumlah Pembatalan dan Tidak Dibatalakan per Hotel:")
for hotel in counts.index:
    print(f"Hotel {hotel} - Not Cancelled: {counts.loc[hotel, 0]}, Cancelled: {counts.loc[hotel, 1]}")

# Show the plot
plt.show()
```

Pada hasil perbandingan tipe hotel di atas, didapat bahwa City Hotel memiliki lebih banyak pemesanan serta pembatalan pemesanan yang lebih tinggi, hal ini Mungkin karena Hotel Resort lebih mahal daripada Hotel City.

## Pertanyaan: Bagaimana pengaruh lead time terhadap pembatalan pemesanan?

```python
canceled = df[df['is_canceled'] == 1]
not_canceled = df[df['is_canceled'] == 0]
```

```python
# Visualisasi Distribusi Lead Time
plt.figure(figsize=(12, 6), facecolor='white')
sns.kdeplot(canceled['lead_time'], label='Cancelled', color='red', shade=True)
sns.kdeplot(not_canceled['lead_time'], label='Not Cancelled', color='green', shade=True)

# Menambahkan Judul dan Label
plt.title('Distribution of Lead Time for Cancelled and Not Cancelled Bookings', size=15)
plt.xlabel('Lead Time (days)', size=12)
plt.ylabel('Density', size=12)
plt.legend()

# Menampilkan plot
plt.show()
```

```python
# Visualisasi Lead Time Rata-rata Berdasarkan Status Pembatalan
plt.figure(figsize=(8, 6), facecolor='white')
sns.barplot(x='is_canceled', y='lead_time', data=df, palette='Set2')

# Menambahkan Judul dan Label
plt.title('Average Lead Time for Cancelled and Not Cancelled Bookings', size=15)
plt.xticks([0, 1], ['Not Cancelled', 'Cancelled'])
plt.xlabel('Booking Status', size=12)
plt.ylabel('Average Lead Time (days)', size=12)

# Menampilkan plot
plt.show()
```

```python
# Output Rata-rata Lead Time
avg_lead_time_cancelled = canceled['lead_time'].mean()
avg_lead_time_not_cancelled = not_canceled['lead_time'].mean()
print(f"Rata-rata Lead Time untuk Booking yang Dibatalkan: {avg_lead_time_cancelled:.2f} hari")
print(f"Rata-rata Lead Time untuk Booking yang Tidak Dibatalkan: {avg_lead_time_not_cancelled:.2f} hari")
```

Dilihat dari data dan visulisasi di atas menunjukkan:

Booking dengan lead time lebih panjang cenderung memiliki peluang lebih besar untuk dibatalkan dan Booking dengan lead time pendek memiliki kemungkinan lebih kecil untuk dibatalkan. dan Ini terlihat dari rata-rata lead time lebih tinggi pada pemesanan yang dibatalkan (105.82 hari) dibandingkan dengan yang tidak dibatalkan (70.44 hari).

## Pertanyaan: Berapa Tarif Harian Rata-rata di city hotel dan resort hotel

```python
resort_hotel= df[df['hotel']=='Resort Hotel']
City_Hotel = df[df['hotel']== 'City Hotel']

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_Hotel = City_Hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8), facecolor='#C38154')
plt.title('Tarif Harian Rata-rata di city hotel dan resort hotel', fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(City_Hotel.index,City_Hotel['adr'],label = 'City Hotel')
plt.legend(fontsize=20)
plt.show()
```

Pada Grafik garis di atas menunjukkan bahwa pada hari-hari tertentu, tarif harian rata-rata untuk city hotel lebih rendah daripada tarif harian Resort hotel, serta juga ada tarif hariannya bahkan lebih rendah lagi.
Ini memungkinkah bahwa pada akhir pekan dan hari libur, tarif Resort hotel akan naik.
